### Handling Missing Values

```python
df.isnull().sum()
```

#### Move Minutes count

```python
missing_MMC = df[df['Move Minutes count'].isnull()]
missing_MMC
```

There are a total of 8 values missing from this column. The way this data is tracked is by calculating all their activity. Tracking for this data is a bit inconsistent since I'm also looking at values with similar metrics however they have records in this column. My solution is to based it on Step count so I give 3 count to the 343, and the rest is 1. However I noticed that the second row has almost all null values and the step count is only 16. This means that this column is the one with missing values of distance, and speed, I'm more inclined to delete this row.

```python
# Delete that one row
null_counts = df[df['Step count'] == 16].isnull().sum(axis=1)
row_to_delete = null_counts[null_counts >= 6].index
df.drop(row_to_delete, inplace=True)

def fill_move_minutes(row):
    if pd.isnull(row['Move Minutes count']):
        if row['Step count'] == 343:
            return 3
        else:
            return 1
    else:
        return row['Move Minutes count']

df['Move Minutes count'] = df.apply(fill_move_minutes, axis=1)
```

#### Activity Duration

There are alot of values missing on the Biking and Running Duration since most of the physical activity I do is walking. I will calculate the total activity duration by summing all 3 values to create a new column for Total Duration.

```python
df['Biking duration (ms)'].fillna(0, inplace=True)
df['Walking duration (ms)'].fillna(0, inplace=True)
df['Running duration (ms)'].fillna(0, inplace=True)

df['Total Duration (ms)'] = df['Biking duration (ms)'] + df['Walking duration (ms)'] + df['Running duration (ms)']
```

#### Heart Points and Minutes

This data records the total Heart Points that a user has obtained from all of their activities. Every data point shows how many Heart Points were determined over a certain period of time.

User's can set a daily Heart Point goal and track their progress. Heart Points can be calculated using:
- heart rate
- Metabolic Equivalent of Task (MET) values
- walking speed

The number of Heart Points earned depends on the intensity of the activity. Use this guide to calculate and write Heart Points:
| Data              | 1 HP (Low-medium intensity activity)                          | 2 HPs (High intensity activity)                                           |
|-------------------|----------------------------------------------------------------|----------------------------------------------------------------------------|
| Heart rate        | 1 minute of activity at >50% maximum heart rate                 | 1 minute of activity at >70% max heart rate                                |
| MET value         | 3-6 MET. For example, rowing or rock climbing                  | >6 MET. For example, HIIT, skiing, volleyball or football                  |
| Walking speed     | 100-130 steps per minute                                       | >130 steps per minute                                                      |


Unfortunately manually calculating for Heart Points and Minutes is a very hard task since I don't have access to some sensor data and it is very hard to do even a approximation of point values. I am forced to drop the columns without any record of Heart Points and Minutes

```python
df.dropna(subset=['Heart Points', 'Heart Minutes'], inplace=True)
```

## Key Insights

### Trend Analysis

```python
daily_metrics = df.groupby('Date').agg({
    'Move Minutes count': 'sum',
    'Calories (kcal)': 'sum',
    'Heart Points': 'sum'
}).reset_index()

plt.figure(figsize=(15, 6))

plt.subplot(311)
plt.bar(daily_metrics['Date'], daily_metrics['Move Minutes count'], color='blue')
plt.title('Move Minutes count per Day')
plt.xlabel('Date')
plt.ylabel('Move Minutes count')

plt.subplot(312)
plt.bar(daily_metrics['Date'], daily_metrics['Calories (kcal)'], color='green')
plt.title('Calories burned per Day')
plt.xlabel('Date')
plt.ylabel('Calories')

plt.subplot(313)
plt.bar(daily_metrics['Date'], daily_metrics['Heart Points'], color='red')
plt.title('Heart Points gained per Day')
plt.xlabel('Date')
plt.ylabel('Heart Points')

plt.tight_layout()
plt.show()
```

```python
df['Heart Points'].describe()
```

```python
# This is already in kcal thus this is just for better visualization
df['Calories (kcal)'] = df['Calories (kcal)'] / 1000.0

daily_metrics = df.groupby('Date').agg({
    'Move Minutes count': 'sum',
    'Calories (kcal)': 'sum',
    'Heart Points': 'sum'
}).reset_index()

# Normalizing the metrics for better visualization
for col in ['Move Minutes count', 'Calories (kcal)', 'Heart Points']:
    daily_metrics[col] = (daily_metrics[col] - daily_metrics[col].min()) / (daily_metrics[col].max() - daily_metrics[col].min())

plt.figure(figsize=(15, 6))
plt.plot(daily_metrics['Date'], daily_metrics['Move Minutes count'], label='Move Minutes')
plt.plot(daily_metrics['Date'], daily_metrics['Calories (kcal)'], label='Calories (kcal)')
plt.plot(daily_metrics['Date'], daily_metrics['Heart Points'], label='Heart Points')
plt.xlabel('Date')
plt.ylabel('Normalized Metrics')
plt.title('Trend Analysis Over Time (Normalized)')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

The gaps are due to the drop rows, which don't have heart points recorded on them. From February to August, my move minutes count and heart points gained per day were in line with each other because during this time period I was still studying at university and was mainly focusing on our thesis, so I did not do much physical activity during this time period. On the other hand, during September, I started a nightly routine of walking for 1 hour and 15 minutes to 2 hours every day, thus this spike in all the metrics. The highest amount of heart points I gained in a day is 120 points, which is close to the weekly target of 150 heart points, according to WHO, so essentially I already completed 80% of a weekly target in one day.

### Correlation/Performance Analysis

```python
relevant_columns = ['Move Minutes count', 'Calories (kcal)', 'Heart Points', 'Heart Minutes', 'Move Minutes count', 'Distance (m)',
                    'Average speed (m/s)', 'Max speed (m/s)', 'Min speed (m/s)', 'Step count']

subset_df = df[relevant_columns]

correlation_matrix = subset_df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Physical Activity Metrics')
plt.show()
```

1. **Move Minutes count and Other Metrics**:
 - Strong Positive Correlations: Move Minutes count shows strong positive correlations with most other metrics: Calories (kcal), Heart Points, Heart Minutes, Distance (m), and Step count. The correlations range from 0.955 to 1.000, indicating a strong positive linear relationship.
 - Highly Correlated Metrics: Step count, Distance (m), Heart Points, Heart Minutes, and Calories (kcal) exhibit high positive correlations with Move Minutes count, indicating that as Move Minutes count increases, these metrics tend to increase consistently.

2. **Calories (kcal) and Other Metrics**: Calories (kcal) shows similar strong positive correlations as Move Minutes count with Heart Points, Heart Minutes, Distance (m), and Step count.

3. **Heart Points and Heart Minutes**:
 - Extremely High Positive Correlation: Heart Points and Heart Minutes show an almost perfect positive correlation of 0.999986-0.999987, indicating an almost perfect linear relationship. An increase in Heart Points correlates nearly identically with an increase in Heart Minutes.

4. **Distance and Step count**: Distance and Step count are almost perfectly positively correlated (0.999607), showing that they tend to increase together consistently.

5. **Speed Metrics (Average speed, Max speed, Min speed)**:
 - Negative Correlations: Speed metrics show negative correlations with most other metrics but are relatively weaker compared to the strong positive correlations observed earlier. As the speed metrics increase, there's a tendency for the Move Minutes count, Calories, Heart Points, and Heart Minutes to decrease.

### Activity Analysis

```python
# Calculate the proportion of each activity duration to Total Duration
df['Biking_proportion'] = df['Biking duration (ms)'] / df['Total Duration (ms)']
df['Walking_proportion'] = df['Walking duration (ms)'] / df['Total Duration (ms)']
df['Running_proportion'] = df['Running duration (ms)'] / df['Total Duration (ms)']

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

sns.scatterplot(data=df, x='Biking_proportion', y='Heart Points', label='Biking', ax=ax1)
sns.scatterplot(data=df, x='Walking_proportion', y='Heart Points', label='Walking', ax=ax1)
sns.scatterplot(data=df, x='Running_proportion', y='Heart Points', label='Running', ax=ax1)
ax1.set_xlabel('Proportion of Activity Duration')
ax1.set_ylabel('Heart Points')
ax1.set_title('Activity Duration vs. Heart Points')
ax1.legend()

duration_cols = ['Biking_proportion', 'Walking_proportion', 'Running_proportion']
sns.boxplot(data=df[duration_cols].dropna(axis=1), ax=ax2)
ax2.set_title('Distribution of Activity Duration Proportions')
ax2.set_ylabel('Proportion')
ax2.set_xlabel('Activity Types')

plt.tight_layout()
plt.show()
```

```python
df['Total_Biking'] = df['Biking duration (ms)'].sum()
df['Total_Walking'] = df['Walking duration (ms)'].sum()
df['Total_Running'] = df['Running duration (ms)'].sum()
total_heart_points = df['Heart Points'].sum()
total_calories = df['Calories (kcal)'].sum()

contribution_heart_points = [
    df['Total_Biking'].sum() / total_heart_points,
    df['Total_Walking'].sum() / total_heart_points,
    df['Total_Running'].sum() / total_heart_points
]

contribution_calories = [
    df['Total_Biking'].sum() / total_calories,
    df['Total_Walking'].sum() / total_calories,
    df['Total_Running'].sum() / total_calories
]

labels = ['Biking', 'Walking', 'Running']

plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.pie(contribution_heart_points, labels=labels, autopct='%1.1f%%')
plt.title('Contribution of Activity Types to Heart Points')
plt.subplot(1, 2, 2)
plt.pie(contribution_calories, labels=labels, autopct='%1.1f%%')
plt.title('Contribution of Activity Types to Calories')
plt.tight_layout()
plt.show()
```

In terms of the proportion of my activity duration and respective activity types, walking contributes the most to the number of heart points I gain and the calories I burn.
