### Quantity Ordered Distribution

```python
def generate_histogram_boxplot(df_column, title_hist=None, title_box=None, bins='auto'):
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

    # Histogram
    sns.histplot(df_column, bins=bins, kde=True, ax=axes[0])
    if title_hist:
        axes[0].set_title(title_hist)

    # Boxplot
    sns.boxplot(x=df_column, ax=axes[1])
    if title_box:
        axes[1].set_title(title_box)

    plt.tight_layout()
    plt.show()
```

```python
generate_histogram_boxplot(df['QuantityOrdered'], title_hist='Distribution of Quantity Ordered', 
                          title_box='Boxplot of Quantity Ordered', bins='auto')
```

```python
def generate_barplot(df, x, y, hue, title=None, xlabel=None, ylabel=None, estimator=None):
    plt.figure(figsize=(20, 5))
    ax = sns.barplot(data=df, x=x, y=y, hue=hue, palette='viridis', estimator=estimator, errorbar=None)
    if title:
        plt.title(title)
    if xlabel:
        plt.xlabel(xlabel)
    if ylabel:
        plt.ylabel(ylabel)
    for p in ax.patches:
        ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='center', fontsize=11, color='black', xytext=(0, 5), 
                    textcoords='offset points')
    plt.show()
```

```python
def generate_sales_trend_plot(x_value, title, xlabel, df):
    sns.set(style="whitegrid", rc={"figure.figsize": (20, 6)})
    sales_by_date = df.groupby([x_value, 'Year'])['Sales'].sum().reset_index()
    sns.lineplot(data=sales_by_date, x=x_value, y='Sales', hue='Year', 
                 palette=sns.color_palette("viridis", len(df['Year'].unique())))
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel("Total Sales")
    plt.grid(True)
    plt.show()
```

#### Time Series Forecasting With Facebook/meta Prophet Model

Prophet is a time series forecasting tool developed by Facebook.It is designed to make accurate time series forecasting accessible to non-experts and has become popular due to its ease of use and ability to handle complex patterns in time series data. It uses a decomposable time series model with three main components: trend, seasonality, and holidays.

```python
from prophet import Prophet

df_FC.columns = ['ds', 'y']
model = Prophet(daily_seasonality=False,
                weekly_seasonality=True,
                yearly_seasonality=True)
model.add_seasonality(name='monthly', period=30.5, fourier_order=5)
model.add_seasonality(name='quarterly', period=91.25, fourier_order=7)
model.fit(df_FC)
```

```python
future = model.make_future_dataframe(periods=214)       #214 days is the amount of days starting from June 1, 2020 to December 31, 2020
sales_forecast = model.predict(future)
```

```python
sales_forecast
```

```python
model.plot(sales_forecast)
plt.show()
```

The black dots represent the actual data points. It can be obsereved that there are some data points concentrated in some order dates this is due to multiple orders being placed on the same date. As seen near at the end of 2020-04 data points stops at May 31, 2020 thus the subsequent value is the forecast made with a set upper and lower boundary.

```python
forecasted_sales = sales_forecast['yhat'][-214:]
forecast_dates = pd.date_range(start='2020-06-01', periods=214)
plt.figure(figsize=(12, 6))
plt.plot(forecast_dates, forecasted_sales, label='Forecasted Sales', color='blue')
plt.legend()
plt.title('Sales Forecast')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

The sales forecast presented interesting results, like having the same characteristics as the previous year's trends of starting the year with low sales and then having high sales at the end of the year. There are 3 points in which it can be observed that the sales reach peaks of 4300+ sales during the times of mid-July, mid-October, and mid-November.

```python
fig = model.plot_components(sales_forecast, weekly_start=1, figsize=(15, 10))
plt.show()
```

The seasonality observed in each trend generated by the forecast aligns consistently with the patterns previously observed in the existing orders.  

### Sales Forecast Performance

```python
from sklearn.metrics import mean_absolute_error

start_date = '2018-01-06'
end_date = '2020-05-31'

# Ranges
actual_sales_range = df_FC[(df_FC['ds'] >= start_date) & (df_FC['ds'] <= end_date)]
forecasted_sales_range = sales_forecast[(sales_forecast['ds'] >= start_date) & (sales_forecast['ds'] <= end_date)]

# Calculate the mean of actual sales and forecasted sales
mean_actual_sales = actual_sales_range['y'].mean()
mean_forecasted_sales = forecasted_sales_range['yhat'].mean()

# Compare mean actual sales with mean forecasted sales
print(f"Mean actual sales for the period {start_date} to {end_date}: {mean_actual_sales}")
print(f"Mean forecasted sales for the period {start_date} to {end_date}: {mean_forecasted_sales}")

# Align the lengths of actual and forecasted sales to match for the calculation
actual_sales_aligned = actual_sales_range['y'].values[:len(forecasted_sales_range)]
forecasted_sales_aligned = forecasted_sales_range['yhat'].values

# Calculate Mean Absolute Error (MAE) between actual and forecasted sales
mae = mean_absolute_error(actual_sales_aligned, forecasted_sales_aligned)
print(f"Mean Absolute Error (MAE) for the period {start_date} to {end_date}: {mae}")
```

- The MAE (2095.07) compared to the mean of actual sales (3553.05) signifies that, on average, the model's prediction errors (in magnitude) are still smaller than the average actual sales value.
- With the MAE (2095.07) being lower than the mean of actual sales (3553.05), it continues to suggest that, on average, the model's predictions are relatively close to the actual sales values for the specified time period.

Comparing the mean of actual sales to the MAE provides a context for understanding the average prediction error of the model relative to the average observed sales. While the MAE is lower than the mean of actual sales, it indicates that, on average, the model's predictions exhibit a relatively good fit to the observed sales data during the analyzed timeframe.
